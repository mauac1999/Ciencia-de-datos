En cuanto a los generos podemos seguir haciendo un analisis y ver cuando venden estos generos en el mundo y los paises

```python
GrupoGenero = data.groupby('Genre')

NorteAmerica_ventas_por_genero = GrupoGenero['NA_Sales'].sum().reset_index()
Europa_ventas_por_genero = GrupoGenero['EU_Sales'].sum().reset_index()
Japon_ventas_por_genero = GrupoGenero['JP_Sales'].sum().reset_index()
mundo_ventas_por_genero = GrupoGenero['Global_Sales'].sum().reset_index()

fig, axes = plt.subplots(2,2, figsize=(15,8), dpi=200, sharey=True)
fig.suptitle('Ventas por genero', fontsize=22)

sns.barplot(ax=axes[0][0], data=NorteAmerica_ventas_por_genero, y='Genre', x='NA_Sales')
sns.barplot(ax=axes[0][1], data=Europa_ventas_por_genero, y='Genre', x='EU_Sales')
sns.barplot(ax=axes[1][0], data=Japon_ventas_por_genero, y='Genre', x='JP_Sales')
sns.barplot(ax=axes[1][1], data=mundo_ventas_por_genero , y='Genre', x='Global_Sales')
plt.tight_layout()


fontsize = 14
labels = [['Norte America', 'Europa'], ['Japon', 'Mundo']]

for x in range(2):
    for y in range(2):
        axes[x][y].set_title(labels[x][y], fontsize=fontsize)
        axes[x][y].set_xlabel('')
```

No por que las empresas saquen mas juegos implica que van a tener un mayor numero de ventas, entonces nos preguntaos ¿Qué consola tuvo la mayor cantidad de juegos y sus ventas en el mundo?

```python
fig, ax = plt.subplots(figsize=(15,5), dpi=200)

sns.countplot(ax=ax, data=data, x='Platform', color='#90A955')
sns.lineplot(ax=ax, data=data,x='Platform', y='Global_Sales', estimator=np.sum, errorbar=None, color='#31572C', marker='o')

plt.ylabel("Ventas globales")
plt.xlabel("Consola")
plt.ylim(0,2500)
plt.grid(True)
plt.title('Juegos por consola vs Ventas globales', fontsize=20)
plt.tight_layout();
```

Por tanto la grafica anterior compara la cantidad de juegos por plataforma y las ventas globales de cada consola. De aqui podemos observar que PS2 y Nitendo DS tenian mayor repertorio de juegos, pero centrandonos en las ventas globales PS2 lidera las listas seguidas por XBOX 360 a pesar de que xbox360 solo saco poco mas de la mitad de juegos que Nintedo DS.
